3. What is Machine Learning?
----
>We can easily blurt out some names of machine learning algorithms such as Logistic Regression, Random Forest, Support Vector Machine and so on. Yes, they are all functions or we can say they are the methods to make our models come into practice. But what is machine learning?

>In 1959, [Arthur Samuel](https://en.wikipedia.org/wiki/Arthur_Samuel) gave out a simple definition of machine learning: Field of study that gives computers the ability to learn without being explicitly programmed.

![checkers](https://raw.githubusercontent.com/RonaldLeoLu/images_for_md/master/CS229/checkers.jpg)

>He designed a program that it can learn how to play games of checkers by training against itself. Because a computer can play thousands of games against itself relatively quickly, the program started to recognize the patterns which led to wins and patterns which led to losses over time. Later amazing things happened that Arthur Samuel's checkers program learned to play checkers even much better than Arthur Sameul himself could. And we can't deny the fact that he didn't program explicitly. The computer just 'trained' itself.

>In 1998, [Tom Mitchel](https://en.wikipedia.org/wiki/Tom_M._Mitchell) gave out a more formal definition about machine learning. He said 'A well-posed learning problem is defined as follow: A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with Experience E.'

>Arthur Samuel's checkers program also fits Tom Mitchel's definition about machine learning. Experience E is the thousands of games the program played against itself; task T is learning how to play checkers better and measure P may be the winning rate or any other evaluative criteria.

> We can simply treat machine learning process as the flow below shows. When we start a task T such as digit recognition, first we need to give  some 'right answers' to the computer which is named as experience E in Tom Mitchel' definition. For example, if you are working on digit recognition, you need to input some pictures of handwrite numbers with artificial labels which shows what the number is in the image. This seems like when meeting new things we have to get rid of the problem what is what. After dealing with the basic rules we can learn further and more complex things. 

![flow](https://raw.githubusercontent.com/RonaldLeoLu/images_for_md/master/CS229/flow.jpg)

>The flow diagram gives us a profile of how machine learning works. The main part is learning. There are lots of steps in the learning part such as data preprocessing, data mining, feature engineering and model training. Then we need the measure P to judge whether our model works well. We will get familiar with loss function later which is commonly used as a kind of measure. If the model doesn't meet our requirement, we'll step into the job of optimization and attempting to tuning our model. Once the model performs as efficient as possible, it will be saved and some works like data visualization or research report will launch.

>First we can see the picture shows the correlation between house price and size in Utopia.(I'm sorry for fudging the data myself in order to express the problem more clearly.) The dataset is consist of two parts, one is the variable _Size_ which gives us the size of the house and the other is _Prize_ which we are interested in. We can treat it as such a problem: when you know the size of a house, how much will you pay for it?

```python
%matplotlib inline
import numpy as np
import random
import matplotlib.pyplot as plt
plt.style.use('ggplot')

x = np.arange(1050, 1130)
y = 0.5 * x + np.random.randint(10, 20, 80)
x1 = np.linspace(1050,1130)
y1 = 0.5 * x1 + 15
plt.figure(figsize=(20,15))
plt.scatter(x,y)
plt.plot(x1,y1)
plt.title('House Price and Size')
plt.xlabel('Size(Inches)')
plt.ylabel('Price(k)')
```

>We often name the variable _Size_ as feature of the dataset and name _Prize_ as a target of our problem. Now what we want to do is just to find a method to show the relation of them. From the picture, we find a line which fits our problem. It can be symbolized as $$y = \theta x + b$$ where x is _Size_ and y is _Price_. 

>Okay, when you get the size of a house, just put it into the formula, make simple calculation and then you can get the house price. This is what we call it _Regression_ problem. And our task is to find the proper value of the parameters $\theta$ and $b$.

>Next problem is referenced from breast cancer dataset. I choose two of the variables, _'Mean radius'_ and _'Mean texture'_. We can find that the whole plot has divided into two parts and if both mean radius and mean texture are known, we are able to tell whether the sample belongs to _'Malignant'_ or _'Benign'_ according to its position in the plot. What we all need to do is to find the margin that correctly separates these two types from one another.
>And this is a typical _Classification_ problem.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import seaborn as sns

bc = load_breast_cancer()
bcdata = pd.DataFrame(bc.data, columns=bc.feature_names)
bcdata['target'] = bc.target
plt.figure(figsize=(20,15))
ax = sns.scatterplot(x='mean radius', y='mean texture', hue='target', data=bcdata, style='target')
ax.legend(fontsize='medium', labels=['Malignant', 'Benign'])
plt.title('Classification Problem')
```

>We may notice that targets of Regression problems are continuous variable while Classification problems' are categorical variable. This is one simple and sometimes useful way to distinguish one from the other.

>Well, these two problems above are similar. They both have some features and a target variable. Our job is to look for an algorithm to predict the sale price or tell the type of cancer. Such works are called _supervised learning_ because we provide a dataset of problems and a target which is sort of a right answer to the learning algorithm or we can say we give the algorithm right answers. 

>An approriate analogy of supervised learning is learning. The process of looking for an algorithm is just like doing homework. 5.Unsupervised Learning
----
>Although it just adds a prefix 'un', unsupervised learning is quite different from supervised learning. If you understand why a learning algorithm is called _supervised_, then you can easily know that unsupervised learning doesn't have the 'right answer'.

>Let's see an example data from NBA. This dataset shows the average assist and rebound per game of each player. We have no idea about the information beyond the data. So first take a look at the scatter plot.

```python
from scripts import avg_preprocess
data = avg_preprocess('./datasets/avg.csv')


dt = data.drop_duplicates('name').head(20)
assist = dt.assist.values
rebound = dt.rebound.values
names = dt.name.values

plt.figure(figsize=(20,15))
plt.scatter(assist, rebound)
plt.title('Data of Some NBA players (part)')
plt.xlabel('assist')
plt.ylabel('rebound')
for label, x, y in zip(names, assist, rebound):
        plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
        plt.xlim(assist.min()+0.00005, assist.max()+0.00005)
        plt.ylim(rebound.min()+0.00005, rebound.max()+0.00005)
```
